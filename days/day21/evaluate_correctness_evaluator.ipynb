```python
import os

from utils import json_load
```

```python
# get evaluator result 
eval_dir = os.path.join('data', 'deliverables', "correctness")
eval_file_path = os.path.join(eval_dir, '2_llama_zh.json')
eval_data = json_load(eval_file_path)  # given qid and get detail
eval_data['1']
```

```python
dataset_dir = os.path.join('data', 'source', 'normalized_exact_match_result')
label_file_path = os.path.join(dataset_dir, '2_llama_zh_evaluation_result.json')
label_data = json_load(label_file_path)
label_data[0]
```

```python
# check by ispass
CNT = 0
labels = []
pred_scores = []

for idx, label in enumerate(label_data):
    qid = label['qid']
    if label['ispass']:
        labels.append(1)
    else:
        labels.append(0)

    correctness = eval_data[qid]
    pred_score = (correctness['score'] - 1 )/ 4
    pred_scores.append(pred_score)

    if (label['ispass'] != correctness['ispass']):
        CNT+=1
        print(f'-----{CNT}')
        print(f"label: {label['ispass']}, 'evaluator: {correctness['ispass']}")
        print("# reference_answer")
        print(correctness['reference_answer'])
        print("# response")
        print(correctness['response'])
        print("# feedback")
        print(correctness['feedback'])
        print("# score")
        print(correctness['score'])
```

```python
# check by score
CNT = 0
labels = []
pred_scores = []

for idx, label in enumerate(label_data):
    qid = label['qid']
    if label['ispass']:
        labels.append(1)
    else:
        labels.append(0)

    correctness = eval_data[qid]
    pred_score = (correctness['score'] - 1 )/ 4
    pred_scores.append(pred_score)

    if int(correctness['score']) != 5:
        correctness['ispass'] = False

    if (label['ispass'] != correctness['ispass']):
        CNT+=1
        print(f'-----{CNT}')
        print(f"label: {label['ispass']}, 'evaluator: {correctness['ispass']}")
        print("# reference_answer")
        print(correctness['reference_answer'])
        print("# response")
        print(correctness['response'])
        print("# feedback")
        print(correctness['feedback'])
        print("# score")
        print(correctness['score'])
```

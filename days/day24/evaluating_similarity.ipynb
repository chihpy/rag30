```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
k8_file_path = os.path.join('data', 'temp', 'baseline_rag_result_semantic.csv')
k3_file_path = os.path.join('data', 'temp', 'baseline_rag_result_R3_semantic.csv')
```

```python
k8df = pd.read_csv(k8_file_path)
k8df.head(3)
```

```python
k8df = k8df[['query', 'reference_answer', 'response', 'similarity_score']].copy()
k8df.columns = ['query', 'reference_answer', 'k8_response', 'k8_similarity_score']
k8df.head(3)
```

```python
k3df = pd.read_csv(k3_file_path)
k3df.head(3)
```

```python
k3df = k3df[['query', 'reference_answer', 'response', 'similarity_score']].copy()
k3df.columns = ['query', 'reference_answer', 'k3_response', 'k3_similarity_score']
k3df.head(3)
```

```python
df = pd.merge(k8df, k3df, how='left', on=['query', 'reference_answer'])
df.head(3)
```

```python
k8_scores = df['k8_similarity_score'].tolist()
k3_scores = df['k3_similarity_score'].tolist()
plt.title('similarity score')
plt.plot(k8_scores)
plt.plot(k3_scores)
```

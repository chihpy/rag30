```python
import os
import json
```

```python
SOURCE_DIR = os.path.join('data', 'source')
os.listdir(SOURCE_DIR)
```

```python
def json_load(file_path):
    with open(file_path, 'r') as f:
        data_dict = json.load(f)
    return data_dict

def json_dump(file_path, data):
    with open(file_path, 'w') as f:
        print("write result to: " + file_path)
        json.dump(data, f, indent=1, ensure_ascii=False)
```

```python
gemma_file_path = os.path.join(SOURCE_DIR, '4_gemma_response.json')
gemma_result = json_load(gemma_file_path)
```

```python
gemma_result['1']
```

```python
dataset_file_path = os.path.join(SOURCE_DIR, 'structured_output_dataset.json')
dataset = json_load(dataset_file_path)
dataset['examples']
```

```python
evaluation_data = []
for example in dataset['examples']:
    reference_answer = example['reference_answer']
    reference_context = example['reference_context']
    qid = reference_answer['qid']
    # get gemma result
    response = gemma_result[qid]['response']
    rv = {
        'reference_answer': reference_answer,
        'reference_context': reference_context,
        'response': response
    }
    evaluation_data.append(rv)
evaluation_data[0]
```

```python
DEST_DIR = os.path.join('data', 'temp')
os.makedirs(DEST_DIR, exist_ok=True)
save_file_path = os.path.join(DEST_DIR, 'gemma_evaluation_data.json')
json_dump(save_file_path, evaluation_data)
```

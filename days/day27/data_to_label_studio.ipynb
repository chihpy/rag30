```python
def make_example(example):
    Ragdialogue = []
    
    query = {
        'author': 'query',
        'text': example['query'],
    }
    Ragdialogue.append(query)
    
    answer = {
        'author': 'reference_answer',
        'text': example['reference_answer'],
    }
    Ragdialogue.append(answer)

    context = {
        'author': 'context',
        'text': example['context']
    }

    Ragdialogue.append(context)

    initial_response = {
        'author': 'initial_response',
        'text': f"ans: {exp['response']['ans']}; feedback: {exp['response']['feedback']}",
    }
    Ragdialogue.append(initial_response)

    if 'response_v2' in example.keys():

        final_response = {
            'author': 'final_response',
            'text': f"ans: {exp['response_v2']['ans']}; feedback: {exp['response_v2']['feedback']}",
        }

        context_v2 = {
            'author': 'context_v2',
            'text': example['context_v2']
        }

        Ragdialogue.append(final_response)
        Ragdialogue.append(context_v2)
    return Ragdialogue
    
```

```python
rvs = []
for exp in data:
    ls_exp = make_example(exp)
    rvs.append({
        'data': {
            'RagResponse': ls_exp
        }
    })

json_dump('rag_response.json', rvs)
```
